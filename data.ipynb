```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
df
```

```python
df.info()
```

```python
df.shape
```

```python
df['director'] = df['director'].fillna('unknown director')
df['cast'] = df['cast'].fillna('unknown cast')
df['country'] = df['country'].fillna('unknown country')
```

```python
df
```

```python
df = df.dropna(subset=['date_added', 'rating', 'duration'])
```

```python
#Cleans the text by removing leading and trailing whitespace.
df['date_added'] = df['date_added'].astype(str).str.strip()
#Converts strings to a proper pandas DateTime object (YYYY-MM-DD format).
df['date_added'] = pd.to_datetime(df['date_added'])
#Fno longer views the column as a calendar date, but as plain text
df['date_added'] = df['date_added'].dt.strftime('%d-%m-%Y')
```

```python
#errors='coerce': This prevents crashes. If a row has corrupt text like "invalid_date", pandas converts it to a safe NaT (Not a Time/missing value) instead of stopping your code.
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year #.dt.year: Extracts just the 4-digit numerical year (e.g., 2026).
df['month_added'] = df['date_added'].dt.month_name() #.dt.month_name(): Extracts the full English name of the month (e.g., July).
df[['date_added', 'year_added', 'month_added']].head()
```

```python
print(df.isnull().sum())
```

```python
#Analyze content types
content_counts = df['type'].value_counts().reset_index()
content_counts.columns = ['Content Type', 'Total Count']
print(content_counts)
```

```python
%pip install nbformat
```

```python
#pie chart
fig_pie = px.pie(content_counts,
    names = 'Content Type',
    values = 'Total Count',
    title = 'Movies vs TV Shows',
    hole = 0.4,
    color_discrete_sequence = ['red','black'])

fig_pie.show()

```

```python
#Top 10 COuntries
top_countries = df['country'].value_counts().reset_index()
top_countries.columns = ['Country', 'Total Content']
```

```python
top_countries = top_countries[top_countries['Country'] != 'unknown country'].head(10)
```

```python
#horizontal bar chart
fig_country = px.bar(top_countries,
                     x = 'Total Content',
                     y = 'Country',
                     orientation='h',
                     title='Top 10 Content producing countries on Netflix',
                     labels={'Total Content':'Number of movies/tv shows','country':'country'},
                     color='Total Content',
                     color_continuous_scale=['black','red'])

fig_country.update_layout(yaxis={'categoryorder':'total ascending'})
fig_country.show()
```

```python
#what kind of stories people are watching
geners_df = df['listed_in'].str.split(', ').explode()
```

```python
top_genres = geners_df.value_counts().reset_index()
top_genres.columns = ['Genre', 'Count']
```

```python
top_genres = top_genres.head(10)
```

```python
#bar chart
fig_genres = px.bar(top_genres,
                    x='Genre',
                    y='Count',
                    title='Top 10 Most popular genres on netflix',
                    labels={'Count':'Number of Titles','Genre':'Genre'},
                    color='Count',
                    color_continuous_scale=['red','black'])
fig_genres.show()
```

Content Duration Analysis

```python
movies_df = df[df['type'] == 'Movie'].copy()
tv_shows_df = df[df['type'] == 'Tv Show'].copy()
```

```python
movies_df['duration_mins'] = movies_df['duration'].str.replace(' min', '').astype(int)
```

```python
fig_movie_dist = px.histogram(
    movies_df,
    x='duration_mins',
    nbins=50,
    title='Distribution of Movie Runtimes on Netflix',
    labels={'duration_mins': 'Duration (Minutes)'},
    color_discrete_sequence=['blue'])
fig_movie_dist.show()
```

```python
tv_shows_df['season_count'] = tv_shows_df['duration'].str.split(' ').str[0].astype(int)
```

```python
tv_season_counts = tv_shows_df['season_count'].value_counts().reset_index()
tv_season_counts.columns = ['Seasons', 'Number of Shows']
tv_season_counts = tv_season_counts.sort_values(by='Seasons')
```

```python
# Plotting the TV Show lifespan
fig_tv_dist = px.bar(
    tv_season_counts,
    x='Seasons',
    y='Number of Shows',
    title='Average Lifespan of Netflix TV Shows (Season Count)',
    labels={'Seasons': 'Number of Seasons', 'Number of Shows': 'Total TV Shows'},
    color='Number of Shows',
    color_continuous_scale=['#221F1F', '#E50914']
)
fig_tv_dist.update_layout(xaxis=dict(tickmode='linear', dtick=1))
fig_tv_dist.show()
```
